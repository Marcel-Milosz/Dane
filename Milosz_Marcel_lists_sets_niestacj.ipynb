01. List manipulation

```python
from random import shuffle

# a) Create a list with 5 fruit names
fruit_names = ["orange", "banana", "strawberry", "pear", "grape"]

# b) Extend the fruit_names list with 2 vegatables
fruit_names.append("cucumber")
fruit_names.append("tomato")

print(fruit_names)
# c) Shuffle the list
shuffle(fruit_names)
print(fruit_names)

# d) Find the position of first vegetable 
for i in fruit_names:
  if i == "cucumber" or i == "tomato":
    print("First veg:", fruit_names.index(i), i)
    break


# e) Remove one fruit
for i in fruit_names:
  if i != "cucumber" or i != "tomato":
    print("Deleting:", fruit_names.index(i), i)
    fruit_names.pop(fruit_names.index(i))
    break



# f) Print the result
print(fruit_names)

# g) Sort the list in alphabetical order
fruit_names_sorted = sorted(fruit_names)

# h) Print each element of a list using a loop
for i in fruit_names_sorted:
  print(i)

# i) Modify above code to be ablet to repeat all above steps 
#    to get the same results each time
```

02. Tuples

```python
first_list = ["a" + str(number) for number in range(1, 10)]
second_list = ["b" + str(number) for number in range(1, 10)]

print(first_list)
print(second_list)

# Pair up above lists
tmp = []
for i in range(0, len(first_list)):
  tupl = (first_list[i], second_list[i])
  tmp.append(tupl)
paired_list = tmp

print(paired_list)


# Using a loop unpack the tuple into variables: first_el, secodn_el
# Print the pair position and variables using f-string.
for i in range(0, len(paired_list)):
  first_el = paired_list[i][0]
  second_el = paired_list[i][1]

  print(f"Infex of {first_el} and {second_el} is equal {i}")
```

3. Sets (unordered & unique data)

```python
fake_names_1 = ['Sherry', 'Mary', 'Matthew', 'Danielle', 'Jeffrey', 'Lauren', 
              'Keith', 'Carlos', 'Monique', 'Laura', 'Jared', 'Valerie', 'Juan', 
              'Christopher', 'Erica', 'Dawn', 'Joshua', 'Brandon', 'Stephanie']

fake_names_2 = ['Andre', 'Anthony', 'Lauren', 'Douglas', 'Jonathan', 'Richard', 
                'Alyssa', 'Vincent', 'Travis', 'Clifford', 'Jerry', 'Justin',
                'Matthew', 'Jared', 'Erica']

# Find the overlapping names in above lists
# Print number of "duplicated" names
duplicates = 0

for i in range(0, len(fake_names_1)):
  if fake_names_1[i] in fake_names_2:
    print("Overlapping name:", fake_names_1[i])
    duplicates += 1

print("Duplicated:", duplicates)

# Find and print the union of above lists
uni = list(set().union(fake_names_1, fake_names_2))
print(uni)

# Find and print the difference between lists
diff = list(set(fake_names_1) - set(fake_names_2))
print(diff)
```
